### 3. Which team has the best attack?

**Answer:**

First of all, to know the team has the best attack, I will select all columns that contribute the attacking of the team, they are stated below:
- attack_scored
- attack_passes
- attack_passes_through
- attack_passes_long
- attack_passes_back
- attack_crosses
- attack_corners_taken
- attack_shots

- attack_goals_headed
- attack_goals_penalty
- attack_goals_box
- attack_goals_outsidebox
- attack_goals_counter
- attack_goals_freekick
- attack_posession
- attack_pass_accuracy

```python
# Get the columns that contains attack keyword
attack_dataset = dataset[['Team', 'attack_scored', 'attack_passes', 'attack_passes_through', 'attack_passes_long',
                          'attack_passes_back', 'attack_crosses', 'attack_corners_taken', 'attack_shots',
                          'attack_shots_on_target', 'attack_goals_headed', 'attack_goals_penalty', 
                          'attack_goals_box', 'attack_goals_outsidebox', 'attack_goals_counter', 
                          'attack_goals_freekick', 'attack_posession', 'attack_pass_accuracy']]
attack_dataset.head()
```

Cited from https://www.guidetofootball.com/tactics/attacking-objectives/ and http://recentfootballnews.com/attacking-an-overview/, the most crucial factors to determine the best attack of the team are:

- **Goal effectiveness**, defines as the conversion of the goal chance into the goal. It can be considered by the total goal and the shot on target
- **Ball possesion**, defines as the combination of dribbling and passing that can prevent the opposite team to has a chance to score a goal
- **Penetration**, defines as the effectiveness of the passing during the game and its accuracy of the passing itself

Therefore, I created some new columns that represents those factors, they are:
- '%goal_converted', defines as the comparation of total goal (sum of 'attack_goals_headed', 'attack_goals_penalty', 'attack_goals_box', 'attack_goals_outsidebox', 'attack_goals_counter', 'attack_goals_freekick') to 'attack_shots_on_target', as the representatives of Goal effectiveness' factor
- '%attack_passes_corner_cross', defines as the comparation of all passing interaction (sum of 'attack_passes_through', 'attack_passes_long', 'attack_passes_back', 'attack_crosses', 'attack_corners_taken') to 'attack_passes', as the representative of Penetration's factor

Thus, to cover all of the factors, I created a new table that contains 4 columns, they are '%goal_converted', '%attack_passes_corner_cross', 'attack_posession', 'attack_pass_accuracy'.

Basically, those 4 columns are required. For '%goal_converted', 'attack_posession', 'attack_pass_accuracy' columns, the bigger the value, the better is the result. Meanwhile for the '%attack_passes_corner_cross', the smaller the value, the better the result.

### Created New Columns of '%goal_converted' and '%attack_passes_corner_cross'

```python
# Create the '%attack_passes_corner_cross' column
attack_dataset['%attack_passes_corner_cross'] = ((attack_dataset['attack_passes_through'] + 
                                     attack_dataset['attack_passes_long'] + 
                                     attack_dataset['attack_passes_back'] +
                                     attack_dataset['attack_crosses'] + 
                                     attack_dataset['attack_corners_taken'])/attack_dataset['attack_passes'])*100
attack_dataset['%attack_passes_corner_cross']
```

```python
# Create the '%goal_converted' column
attack_dataset['total_goal'] = (attack_dataset['attack_goals_headed'] + 
                                 attack_dataset['attack_goals_penalty'] + 
                                 attack_dataset['attack_goals_box'] + 
                                 attack_dataset['attack_goals_outsidebox'] + 
                                 attack_dataset['attack_goals_counter'] + 
                                 attack_dataset['attack_goals_freekick'])
attack_dataset['%goal_converted'] = (attack_dataset['total_goal']/attack_dataset['attack_shots_on_target'])*100
attack_dataset['%goal_converted']
```

### Generate The New Table

```python
# Create a new variable of selected columns
attack_dataset_fix = attack_dataset[['Team', '%goal_converted', 'attack_posession', '%attack_passes_corner_cross', 'attack_pass_accuracy']]
attack_dataset_fix
```

### Calculate The Mean of selected columns

```python
# Count the mean value
attack_dataset_fix['%mean_attacking'] = attack_dataset_fix[['%goal_converted', 'attack_posession', 'attack_pass_accuracy']].mean(axis=1)
attack_dataset_fix.head()
```

```python
# Select only the final columns
attack_dataset_fixed = attack_dataset_fix[['Team', '%mean_attacking', '%attack_passes_corner_cross']]
attack_dataset_fixed.head()
```

```python
# Create scatter plot
fig, ax = plt.subplots(figsize=(20,15))

sns.scatterplot(data = attack_dataset_fixed, x = "%mean_attacking", 
                y = "%attack_passes_corner_cross", s = 200, ax = ax,
                hue = attack_dataset_fixed['Team']).set(title = 'Mean_Attacking vs Attacking_Passess_Corner')
```
